## 1. Total Revenue/Sales

```python
import pandas as pd
```

```python
total_sales = (df['price'] * df['quantity']).sum()
print(f'Total Revenue: ${total_sales:.2f}')
```

## 2. Total Quantity Sold

```python
total_quantity = df['quantity'].sum()
print(f'Total Quantity Sold: {total_quantity}')
```

## 3. Total Orders

```python
total_orders = df['order_id'].nunique()
print(f'Total Orders: {total_orders}')
```

## 4. Number of Pizza Types

```python
pizza_types_count = df['pizza_type_id'].nunique()
print(f'Pizza Types Sold: {pizza_types_count}')
```

## 5. Average Pizza Price

```python
average_price = df['price'].mean()
print(f'Average Pizza Price: ${average_price:.2f}')
```

## 6. Peak Hours of Sales

```python
df['hour'] = pd.to_datetime(df['time'], format='%H:%M:%S').dt.hour
sales_by_hour = df.groupby('hour')['quantity'].sum().sort_values(ascending=False)
print(sales_by_hour.head())
```
